```python
from Matrix import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Scatterplot with separate X and Y matrices

```python
x = Matrix([1, 2, 3, 4, 5])
y = Matrix([2, 4, 1, 3, 5])

plt.scatter(x.data(), y.data())
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

### Bar Plot with single Matrix

```python
import matplotlib.pyplot as plt

matrix = Matrix([['A', 'B', 'C', 'D'], [10, 15, 7, 12]])

plt.bar(matrix.data()[0], matrix.data()[1])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot Example")
plt.show()
```

#### Line Chart from 2D Matrix

```python
matrix = Matrix([[1, 2, 3, 4, 5], [2, 4, 1, 3, 5]])

sns.lineplot(x=matrix.data()[0], y=matrix.data()[1])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Chart Example")
plt.show()
```

#### Heatmap example

```python
matrix = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])

sns.heatmap(matrix.data(), annot=True, cmap='coolwarm')
plt.title("Heatmap Example")
plt.show()
```
